```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_stat = pd.read_csv("status.csv")

df_stat.head()
```

```python
# convert time column to datetime
df_stat["time"] = pd.to_datetime(
    df_stat["time"],
    format="%Y/%m/%d %H:%M:%S",
    errors="coerce"
)

```

```python
start_time = df_stat["time"].min()
end_time = df_stat["time"].max()
print("Start time:", start_time)
print("End time:  ", end_time)
```
