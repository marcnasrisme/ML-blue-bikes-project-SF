## Modeling Realized vs. Latent Demand

In this project, we define the hourly net demand at station $i$ during hour $t$ as:

$$
\text{net\_demand}_{i,t} = b_{i,t} - b_{i,t+1}.
$$

Here, $b_{i,t}$ denotes the number of bikes available at the start of hour $t$.  
This captures the realized net flow of bikes (departures minus arrivals) during that hour.

It is important to emphasize that this quantity does **not** represent the true latent demand.  
When a station is completely full or completely empty, attempted arrivals or departures may be **censored** because riders are unable to dock or pick up a bike. As a result, the observed inventory may fail to reflect the underlying unmet demand.

For example, if a station is full at both 6:00 and 7:00, the formula above yields:

$$
\text{net\_demand}_{i,t} = C_i - C_i = 0,
$$

even though several riders may have attempted (and failed) to return their bikes during that period.

Because the dataset contains only inventory snapshots (bikes and docks available), the true latent demand cannot be recovered in these censored scenarios. In this project, we therefore interpret $\text{net\_demand}_{i,t}$ as **realized** demand and acknowledge that it systematically underestimates true demand whenever stations are saturated.

To mitigate this limitation, we focus our predictive modeling on periods where stations operate away from their capacity bounds (i.e., not near empty or full), where censoring is less severe. Moreover, in the optimization stage, buffer constraints can be incorporated into the MIO formulation to reduce the risk of operating stations too close to empty or full states — precisely the conditions where censoring issues are most problematic.

## For later

We evaluate our optimized initial allocations using historical days from the test period. For a given day, we treat the realized net demands at each station and hour as an exogenous disturbance sequence, and simulate inventory trajectories under our optimized starting levels. This allows us to compute the resulting number of empty/full station events and compare it to the historical baseline. This approach implicitly assumes that short-term demand is independent of the initial allocation; in reality, availability may affect user choices (e.g., walking to a nearby station), but such counterfactual demand responses are not identifiable from the available data. We therefore focus on this standard exogenous-demand approximation.

### Including some lags because why not

```python
# Including previous day, same hour lag for net demand
df_master = df_master.sort_values(["station_id", "hour_of_day", "date"])

df_master["net_demand_prev_day_same_hour"] = (
    df_master.groupby(["station_id", "hour_of_day"])["net_demand"]
             .shift(1)
)

# Now the moving average of net demand for the past 3 days, same hour
df_master["net_demand_ma_3days_same_hour"] = (
    df_master.groupby(["station_id", "hour_of_day"])["net_demand"]
      .rolling(window=3, min_periods=1)   # last 3 days at that hour
      .mean()
      .reset_index(level=[0,1], drop=True)
)

df_master.head()
```

### Train/Test Split

Given we only have 6 months of data, and we want to avoid data leakage, we are gonna holdout february for our test set, and cross validate on the previous months (Aug to Jan)

```python
import numpy as np

# Sort by time (as you already do)
df_master = df_master.sort_values(["date", "hour_of_day", "station_id"]).reset_index(drop=True)

# Add a stable row id *after* sorting
df_master["obs_id"] = np.arange(len(df_master))

print("obs_id in columns?", "obs_id" in df_master.columns)
```

```python
from sklearn.model_selection import TimeSeriesSplit
import numpy as np
import pandas as pd

# ---------------------------------------
# 0. Work on a COPY of df_master
# ---------------------------------------
# assumes df_master already exists in memory with all features

dfm = df_master.copy()

# replace missing in the "same hour previous day" feature by 0
dfm["net_demand_prev_day_same_hour"] = dfm["net_demand_prev_day_same_hour"].fillna(0)

# (optional but consistent): also zero any missing in the 3-day MA lag
dfm["net_demand_ma_3days_same_hour"] = dfm["net_demand_ma_3days_same_hour"].fillna(0)

# quick sanity check on the copy
print("NaNs in net_demand_prev_day_same_hour (copy):",
      dfm["net_demand_prev_day_same_hour"].isna().sum())
print("NaNs in net_demand_ma_3days_same_hour (copy):",
      dfm["net_demand_ma_3days_same_hour"].isna().sum())

# ---------------------------------------
# 1. Sort + create obs_id on the copy
# ---------------------------------------
dfm = dfm.sort_values(["date", "hour_of_day", "station_id"]).reset_index(drop=True)
dfm["obs_id"] = np.arange(len(dfm))

dfm["date"] = pd.to_datetime(dfm["date"])

# ---------------------------------------
# 2. Chronological train / test split
# ---------------------------------------
cutoff_date = pd.Timestamp("2014-01-31")  # everything AFTER this is test (i.e., Feb)

df_train = dfm[dfm["date"] <= cutoff_date].copy()
df_test  = dfm[dfm["date"] >  cutoff_date].copy()

print("Train dates:", df_train["date"].min(), "→", df_train["date"].max())
print("Test  dates:", df_test["date"].min(),  "→", df_test["date"].max())

# 🔹 Only drop rows where the TARGET is missing (NOT zeros)
before_test = len(df_test)
df_test = df_test[~df_test["net_demand"].isna()].copy()
print("Dropped rows with missing net_demand in TEST:", before_test - len(df_test))

# ---------------------------------------
# 3. Features and target
# ---------------------------------------
feature_cols = [
    "hour_of_day", "day_of_week", "is_weekend", "month", "is_winter",
    "lat", "long", "capacity",
    "bikes_available", "docks_available", "load_factor",
    "mean_temperature_f", "mean_wind_speed_mph", "precipitation_inches",
    "net_demand_prev_day_same_hour", "net_demand_ma_3days_same_hour",
]

target_col = "net_demand"

X_train = df_train[feature_cols].values
y_train = df_train[target_col].values

# ---------------------------------------
# 4. Time-series CV on TRAIN ONLY
#    Save folds into ONE CSV
# ---------------------------------------
tscv = TimeSeriesSplit(n_splits=5)

fold_rows = []  # will hold (obs_id, fold, role)

for fold, (train_idx, val_idx) in enumerate(tscv.split(X_train), start=1):
    print(f"Fold {fold}")
    print("  Train range:", train_idx[0], "→", train_idx[-1])
    print("  Val   range:", val_idx[0],   "→", val_idx[-1])

    train_obs_ids = df_train.iloc[train_idx]["obs_id"].to_numpy()
    val_obs_ids   = df_train.iloc[val_idx]["obs_id"].to_numpy()

    for oid in train_obs_ids:
        fold_rows.append({"obs_id": int(oid), "fold": fold, "role": "train"})
    for oid in val_obs_ids:
        fold_rows.append({"obs_id": int(oid), "fold": fold, "role": "val"})

cv_folds_df = pd.DataFrame(fold_rows)
cv_folds_df.to_csv("bluebikes_ts_cv_folds.csv", index=False)

# ---------------------------------------
# 5. Save ML tables (this is what Julia will read)
# ---------------------------------------
df_train_ml = df_train[["obs_id"] + feature_cols + [target_col]].copy()
df_test_ml  = df_test[["obs_id"]  + feature_cols + [target_col]].copy()

print("Train ML shape:", df_train_ml.shape)
print("Test  ML shape:", df_test_ml.shape)

df_train_ml.to_csv("bluebikes_train_ml.csv", index=False)
df_test_ml.to_csv("bluebikes_test_ml.csv", index=False)

# ---------------------------------------
# 6. Explicit X_test and y_test CSVs (Feb)
# ---------------------------------------
X_test_df = df_test[["obs_id"] + feature_cols].copy()
y_test_df = df_test[["obs_id", target_col]].copy()

X_test_df.to_csv("bluebikes_X_test.csv", index=False)
y_test_df.to_csv("bluebikes_y_test.csv", index=False)

print("Saved:")
print("  bluebikes_train_ml.csv")
print("  bluebikes_test_ml.csv")
print("  bluebikes_X_test.csv")
print("  bluebikes_y_test.csv")
print("  bluebikes_ts_cv_folds.csv")
```
